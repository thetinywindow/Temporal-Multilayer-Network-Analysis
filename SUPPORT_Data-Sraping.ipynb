```python
#iter through rows 
a=[]  
for index, row in df_ref.iterrows():
    #print(row.T)
    a.append(row.T)

#check length according to input of papers
len(a)
```

```python
#transform to dataframe
def_ref_names = pd.DataFrame(a)
#def_ref_names

#fill missing values with 'na'
def_ref_names = def_ref_names.replace(r'^\s*$', np.nan, regex=True)
def_ref_names = def_ref_names.fillna('na')

#output
def_ref_names.head(1)
```

```python
import re
```

```python
#transform reference dataframe to list
a = def_ref_names.values.tolist()
```

```python

#generate IDs for each reference; enumerating through list
for lst in a:
    for ind, item in enumerate(lst):
        lst[ind] = lst[ind].split(',', 1)[0]
        #lst[ind] = ''.join([i for i in lst[ind] if not i.isdigit()])
        #lst[ind] = ''.join(e for e in lst[ind] if e.isalnum())
        lst[ind] = re.sub(r'[^a-zA-Z ]',r'',lst[ind])
        if lst[ind] == ' Anonymous':
            lst[ind] = 'na'

#transform list to dataframe
def_ref_names = pd.DataFrame(a)

#output reference id dataframe
def_ref_names.head()
```

```python
#transform reference dataframe to list
a = def_ref_names.values.tolist()

#generate IDs for each reference; enumerating through list
for lst in a:
    for ind, item in enumerate(lst):
        lst[ind] = id(item)

#transform list to dataframe
def_ref_id = pd.DataFrame(a)

#output reference id dataframe
def_ref_id.head()
```

```python
#transform all values from int to string
def_ref_id = def_ref_id.applymap(str)

#combine all columns into one seperated by comma
combine_col = def_ref_id.apply(lambda x: ', '.join(filter(None, x)), axis=1)
df_ref_id_combined = pd.DataFrame(combine_col, columns=['REFERENCE_LIST_ID'])

#output combines reference by author
df_ref_id_combined.head(2)
```

```python
df_combine = pd.DataFrame(df['PT'])

df_test = pd.concat([df_combine.reset_index(drop=True),df_ref_id_combined.reset_index(drop=True)], axis=1)
df_test
```

```python
#generate dataframe for author references (each reference in seperate column)
df_author = df_test['PT'].str.split(';', 10, expand=True)

author_list = df_author.values.tolist()

#generate IDs for each reference; enumerating through list
for lst in author_list:
    for ind, item in enumerate(lst):
        lst[ind] = id(item)  

#transform list to dataframe
def_author_id = pd.DataFrame(author_list) #, columns=['AUTHOR_LIST_ID'])

#transform all values from int to string
def_author_id = def_author_id.applymap(str)

#combine all columns into one seperated by comma
combine_col = def_author_id.apply(lambda x: ', '.join(filter(None, x)), axis=1)
df_author_id_combined = pd.DataFrame(combine_col, columns=['AUTHOR_LIST_ID'])

#output combines reference by author
df_author_id_combined.shape, df_author_id_combined.head(2)
```

```python
df_author_network = pd.concat([df_test.reset_index(drop=True),df_author_id_combined.reset_index(drop=True)], axis=1)
#df_author_network
```
